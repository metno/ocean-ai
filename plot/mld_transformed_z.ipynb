Calculating the mld depth using a transformation from s_levels to z_rho. This results in more bumpy plots, which is why a interpolation might be a better fit for this type of calculation. Even though, this notebook is to visualize the differences and compare these two methods of calculating the mld. 

```python
from script_mld import dens
from script_mld import transformation
import xarray as xr
import matplotlib.pyplot as plt 
```

```python
ds_2024 = xr.open_dataset(f'/lustre/storeB/project/fou/hi/foccus/datasets/symlinks/norkystv3-hindcast/2024/norkyst800-20240330.nc').isel( time = 0, X = 360, Y = 760)
temp = ds_2024.temperature.transpose() 
salinity = ds_2024.salinity.transpose() 
pot_dens = dens(temp, salinity)
pot_dens = pot_dens.values
transformation(ds_2024)
z_rho = ds_2024.z_rho.values
```

```python
plt.plot(pot_dens, z_rho, color = "m")
plt.xlabel("Potential Density")
plt.ylabel("Z_rho using nearest neighbour")
plt.title("Vertical plot of potential density and z_rho transformed")
```
