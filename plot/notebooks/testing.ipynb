```python
norkyst3_ts_data = norkyst3.salinity.isel(time=-1)
havbris_ts_data  = havbris.salinity.isel(time=-1)

plt.figure(figsize=(20,6))
plt.imshow(norkyst3_ts_data,vmin=33,vmax=35)
plt.colorbar()
plt.title(f'Norkyst3, timestep=1, date={havbris_ts_data.time.values}',fontsize=20)
plt.show()
plt.figure(figsize=(20,6))
plt.imshow(havbris_ts_data,vmin=33,vmax=35)
plt.colorbar()
plt.title(f'Havbris, timestep=1, date={havbris_ts_data.time.values}',fontsize=20)
plt.show()
diff = havbris_ts_data-norkyst3_ts_data
plt.figure(figsize=(20,6))
plt.imshow(diff,cmap='bwr',vmin=-0.5,vmax=0.5)
plt.colorbar()
plt.title(f'Difference (Havbris-Norkyst3), timestep=1, date={havbris_ts_data.time.values}',fontsize=20)
plt.show()
```
